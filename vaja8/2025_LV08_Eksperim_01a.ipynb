```python
# 2.1


import pandas as pd 

# Nalozi podatke iz lokalne datoteke
fn = 'SUS_Vprasalnik_2025.xlsx'


df = pd.read_excel(fn)
#df
```

```python
# Izracunaj stolpce Odd, Even, SUS
df['Odd'] = df['Q1']+df['Q3']+df['Q5']+df['Q7']+df['Q9'] - 5
df['Even'] = 25 - (df['Q2']+df['Q4']+df['Q6']+df['Q8']+df['Q10'])
df['SUS'] = (df['Odd'] + df['Even'] )*2.5
```

```python
# Izpisi dataframe
df.head(10)

#print(df)
```

```python
# Skupine
S1 = df['S']=='Google'
S2 = df['S']=='ChatGPT'

# Preveri
df[S2]
```

```python
# Izris po skupinah
df.boxplot(by='S', column=['Even','Odd'])
```

```python
# Izris SUS po skupinah
df.boxplot(by='S', column=['SUS'])
```

```python
# Histogram SUS po napravah
df.hist(by='S', column=['SUS'], bins=10)
```

```python
# liliefors : Ali je porazdelitev normalna, p > 0.05
from statsmodels.api import stats

# Naprava 1
S1lil = stats.diagnostic.lilliefors(df[S1]['SUS'])
print(S1lil)
print("Alexa > Normalna porazd p: ", S1lil[1])

# Naprava 2
Slil2 = stats.diagnostic.lilliefors(df[S2]['SUS'])
print(Slil2)
print("Google > Normalna porazd p: ", Slil2[1])


```

```python
# TTest ničelna hipoteza: ni razlike med sistemi, p>0.05
from scipy import stats

# Preveriti moramo, da je enaka dolžina podatkov
print(len(df[S1]['SUS']), len(df[S2]['SUS']))

Nlast = 6

results = stats.ttest_rel(df[S1]['SUS'][0:Nlast], df[S2]['SUS'][0:Nlast])
print("TTest: ", results)
print("p: ", results.pvalue)
```

```python
# Cronbach alpha
import pingouin as pg

data1 = pd.DataFrame(df, columns=['Q1','Q3','Q5','Q7','Q9'])
print(data1)
print(pg.cronbach_alpha(data=data1))

# Za drugo skupino stolpcev Q2, Q4, ..
data2 = pd.DataFrame(df, columns=['Q2','Q4','Q6','Q8','Q10'])
print(data2)
print(pg.cronbach_alpha(data=data2))
```

## SUISQ Analiza

```python
import pandas as pd 

# Nalozi podatke iz lokalne datoteke
fn = 'SUISQR_24a.xlsx'

df = pd.read_excel(fn)
df
```

```python
# Izracunaj faktorje UGO, CSB, SC, V:
df['UGO']=(df['Q1']+df['Q2']+df['Q3']+df['Q4'])/4.0
df['CSB']=(df['Q5']+df['Q6']+df['Q7']+df['Q8'])/4.0
df['SC']=(df['Q9']+df['Q10']+df['Q11'])/3.0
df['V']=(df['Q12']+df['Q13']+df['Q14'])/3.0
```

```python
df.head(10)
```

```python
# Skupine 
S1 = df['S']=='Alexa'
S2 = df['S']=='Google'
```

```python
# Boxplot po faktorjih
df.boxplot(by='S', column=['UGO','CSB','SC','V'])
```

```python
# liliefors : Ali je porazdelitev normalna, p > 0.05, za vse faktorje
from statsmodels.api import stats
S1lil = stats.diagnostic.lilliefors(df[S1]['SC'])
print("SC Normalna porazd p: ", S1lil[1])

```

```python
# Histogram verbosity
df.hist(column=['UGO'], bins = 10, by='S')
df.hist(column=['CSB'], bins = 10, by='S')
```

```python
# TTest ničelna hipoteza: ni razlike med sistemi, p>0.05
from scipy import stats
len(df[S2]['UGO']), len(df[S1]['UGO'])

# UGO
print('UGO', stats.ttest_rel(df[S1]['UGO'][0:21], df[S2]['UGO'][0:21]))
# CSB
print('CSB', stats.ttest_rel(df[S1]['CSB'][0:21], df[S2]['CSB'][0:21]))
# SC
print('SC', stats.ttest_rel(df[S1]['SC'][0:21], df[S2]['SC'][0:21]))
# V
print('V', stats.ttest_rel(df[S1]['V'][0:21], df[S2]['V'][0:21]))
```
